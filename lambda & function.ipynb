```python
data = {
    'Nama': ['Andi', 'Budi', 'Cici', 'Bambang', 'Luthfi'],
    'Umur': [25, 30, 22, 31, 19],
    'Gaji Bersih': [5000, 6000, 4500, 6000, 10000],
}

df = pd.DataFrame(data)
```

Pertanyaan no 1: 
buatlah lopp for yang akan mengiterasi melalui DataFrame untuk mengecek usia tiap karyawan . 
jika usia diatas 30 tahun, tambahkan bonus sebanyak 5%.

```python
def persentase_bonus(umur):
    bonus = .05

    return bonus if umur > 30 else 0

df['Persentase Bonus'] = df['Umur'].apply(persentase_bonus)
df['Bonus'] = df['Gaji Bersih'] * df['Persentase Bonus']

df
```

Pertanyaan no 2:
setelah pemberian bonus, tampilkan DataFrame yang sudah diperbaharui, lalu hitung jumlah total gaji sebelum dan sesudah bonus

```python
# pertanyaan no 2

df['Gaji Sebelum Bonus'] = df['Gaji Bersih']
df['Gaji Setelah Bonus'] = df['Gaji Bersih'] + df['Bonus']

df
```

Pertanyaan no 3:
buat ringkasan hasil, termasuk daftar karyawan yang menerima bonus dan yang tidak menerima bonus

```python
# pertanyaan no 3

print('Daftar pegawai yang menerima bonus')
display(df.query('Bonus > 0')[['Nama', 'Bonus', 'Gaji Sebelum Bonus', 'Gaji Setelah Bonus']])

print('Daftar pegawai yang tidak menerima bonus')
display(df.query('Bonus == 0')[['Nama', 'Bonus', 'Gaji Sebelum Bonus', 'Gaji Setelah Bonus']])
```
