```python
import pandas as pd
data = {'nama': ['andi', 'budi', 'cici'],
'Umur' : [25, 30, 22], 
'Gaji' : [5000,6000,4500]}

df = pd.DataFrame(data)
def kategori_umur(umur):
    if umur <25 :
        return 'Muda'
    elif 25 <= umur <35:
        return 'Dewasa'
    else :
        return 'Lansia'

df['Kategori_Umur'] = df['Umur'].apply(kategori_umur)
#kolom baru di dataframe variabel df
df['Gaji_Bersih'] = df['Gaji'].apply(lambda x:0.9 * x)
print(df)
```

```python
import pandas as pd
file_path = "C:\\Users\\Silvy Nur Azkia\\Downloads\\banjir.xlsx"
df = pd.read_excel(file_path)
print(df)

df = pd.DataFrame(df)
def Status_banjir(kategori_banjir):
    if kategori_banjir <1 :
        return 'aman'
    elif 1 <= kategori_banjir <5:
        return 'siaga'
    else :
        return 'awas'

df['status'] = df['jumlah_banjir'].apply(Status_banjir)
print(df)
```
